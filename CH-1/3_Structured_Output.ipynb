Guiding in Prompts

```python
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0)
```

```python
from langchain_core.prompts import PromptTemplate

result = llm.invoke("Tell me a joke. Generate the prompt in key value pairs. Key should be setup and punchline.")
result.content
```

Using pydantic models

```python
from pydantic import BaseModel, Field

class llm_output(BaseModel):
    setup: str = Field(description="The setup of the joke")
    punchline: str = Field(description="The punchline of the joke")
```

```python
llm_structured = llm.with_structured_output(llm_output)
result = llm_structured.invoke("Tell me a joke")
print(result)
```
