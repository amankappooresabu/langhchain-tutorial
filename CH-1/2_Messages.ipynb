```python
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0)

my_messages = [
    SystemMessage(content="You are a serious assistant."),
    HumanMessage(content="What is the capital of France?"),
]

llm.invoke(my_messages).content
```

```python
from langchain_core.prompts import PromptTemplate

user_input = input("What topic do you want a joke about? ")

dynamic_prompt = PromptTemplate.from_template("tell me a joke about {topic}")

ready = dynamic_prompt.invoke({"topic": user_input})

llm.invoke(ready).content
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt_template = ChatPromptTemplate.from_messages(
    [
        ("system","You are a serious assistant."),
        ("user", "Tell me a joke about {topic}."),
    ]
)

user_input = input("What topic do you want a joke about? ")

ready = prompt_template.invoke({"topic": user_input})

llm.invoke(ready).content
```
